```python
query="select * from banking.customer"
```

```python
df=pd.read_sql(query,conn)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bins = [0,100000,300000,float('inf')]
labels=['low','Medium','High']
df['Income value'] = pd.cut(df['Estimated Income'],bins=bins,labels=labels,right=False)
```

```python
df['Income value'].value_counts().plot(kind='bar')
```

```python
bins = [15,30,50,float('inf')]
labels=['low','Medium','High']
df['Age category'] = pd.cut(df['Age'],bins=bins,labels=labels,right=False)
```

```python
df['Age category'].value_counts().plot(kind="bar")
```

```python
Examine the distribution of unique cataegories in categorical columnsdisplay
```

```python
categorical_cols=df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure","Income value", "Loyalty Classification", "Properties Owned", "Risk Weighting"]].columns
for col in categorical_cols:
    print(f"value counts for {col}:")
    display(df[col].value_counts())
```

```python
UNIVARIATE ANALYSIS
```

```python

for predictor in df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income value"]].columns:
    plt.figure(predictor) 
    sns.countplot(data=df, x=predictor,hue="GenderId")
    plt.title(f"Count plot of {predictor}")
    plt.show()
```

```python
Histplot of valuecounts 
```

```python
for col in categorical_cols:
    if col == "Occupation":
       continue
    plt.figure(figsize=(8,4))
    sns.histplot(data=df,x=col)
    plt.show()
```

```python
Numeriacal Analysis
```

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']
plt.figure(figsize=(18,13))
for i,col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()
    
```

```python
HEATMAP
```

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']
plt.figure(figsize=(18,13))
correlation_matrix = df[numerical_cols].corr()
sns.heatmap( df[numerical_cols].corr(),annot=True,cmap="coolwarm",linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```
